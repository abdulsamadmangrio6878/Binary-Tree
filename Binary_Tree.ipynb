```python
class TREE_NODE:
    def __init__(a,e):
        a.value = e
        a.L = None
        a.R = None

def insert(n,e):
    if n is None:
        return TREE_NODE(e)
    if e > n.value:
        n.R = insert(n.R,e)
    else:
        n.L = insert(n.L,e)
    return n

root = None
root = insert(root,12)
root = insert(root,3)
root = insert(root,14)
root = insert(root,6)
root = insert(root,30)
root = insert(root,8)
root = insert(root,2)
root = insert(root,11)
root = insert(root,32)
root = insert(root,7)
root =insert(root,19)
```

```python
class BINARY_TREE_NODE:
    def __init__(a):
        pass

    def size(a,n):
        if n is not None:
            return n
        s = 1 + a.size(n.R) + a.size(a.L)
        return s
    
    def height(a,n):
        if n is None:
            return -1
        h = 1 + max(a.height(n.L),a.height(a.R))
        return h
    
    def minimum(a,n):
        if n.L is None:
            return n
        return a.minimum(n.L)
    
    def maax(a,n):
        if n.R == None:
            return n
        return a.maax(n.R)
    
    def delete(a,n,e):
        if n is None:
            return n
        if e > n.value:
            n.R = a.delete(n.R,e)
        elif e < n.value:
            n.L = a.delete(n.L,e)
            
        else:
            if n.R is None:
                temp = n.L
                n = None
                return temp
            elif n.L is None:
                temp = n.R
                n = None
                return temp
            
            temp = a.minimum(n.R)
            n.value = temp.value
            n.R = a.delete(n.R,temp.value)

        return n


    def pre_order(a,n):
        if n is None:
            return "Error"
        print(n.value,end=" ")
        a.pre_order(n.L)
        a.pre_order(n.R)

    def In_Order(a,n):
        if n is None:
            return "Error"
        a.In_Order(n.L)
        print(n.value,end=" ")
        a.In_Order(n.R)

    def Post_Order(a,n):
        if n is None:
            return "Error"
        a.Post_Order(n.L)
        a.Post_Order(n.R)
        print(n.value,end= " ")

bt = BINARY_TREE_NODE()
print("pre order VLR")
bt.pre_order(root)
print()
print(bt.delete(root,11).value)
print("In_Order LVR")
bt.In_Order(root)
bt.size(root)


# bt.height(root)
```
